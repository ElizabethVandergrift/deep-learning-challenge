## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
# Drop the 'EIN' and 'NAME' columns and assign the result back to the DataFrame
application_df = application_df.drop(columns=['EIN', 'NAME'])

# Display the first few rows of the updated DataFrame
application_df.head()
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
value_counts=application_df['APPLICATION_TYPE'].value_counts()
print(value_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Define the threshold
threshold = 200  # Replace values below this count with "Other"

# Calculate value counts
value_counts = application_df['APPLICATION_TYPE'].value_counts()

# Identify application types to replace
application_types_to_replace = value_counts[value_counts < threshold].index

# Replace in dataframe
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(application_types_to_replace, "Other")

# Check to make sure replacement was successful
print(application_df['APPLICATION_TYPE'].value_counts())

```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
value_counts=application_df['CLASSIFICATION'].value_counts()
print(value_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
print(value_counts[value_counts > 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Define the threshold

threshold = 700

# Identify classifications to replace with "Other"
classifications_to_replace = value_counts[value_counts < threshold].index

# Replace in dataframe
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(classifications_to_replace, "Other")

# Display the updated value counts to check if the replacement was successful
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
value_counts=application_df['ASK_AMT'].value_counts()
print(value_counts)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scale ASK_AMT using Min-Max scaling
scaler = MinMaxScaler()
application_df['ASK_AMT'] = scaler.fit_transform(application_df[['ASK_AMT']])
```

```python
# Convert categorical data to numeric with pd.get_dummies and ensure all data is of type int
application_dummies = pd.get_dummies(application_df[["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "STATUS", "INCOME_AMT", "SPECIAL_CONSIDERATIONS"]]).astype(int)

# Step 3: Combine the scaled ASK_AMT column with the one-hot encoded data and the target column
application_df_scaled = pd.concat([
    application_df.index.to_frame(),  # Keep the index as a column
    application_dummies,              # One-hot encoded categorical data
    application_df["ASK_AMT"],        # Scaled ASK_AMT column
    application_df["IS_SUCCESSFUL"]   # Target column
], axis=1)

# Display the combined DataFrame
application_df_scaled.head()
```

```python
application_df_scaled.columns
```

```python
#application_df_scaled = application_df_scaled.drop(columns=['APPLICATION_TYPE_Other', 'CLASSIFICATION_Other'])
application_df_scaled.head()
```

```python
# Split our preprocessed data into our features and target arrays
# The y variable should focus on the target column
y = application_df_scaled['IS_SUCCESSFUL']

# The X variable should include all features except the target
X = application_df_scaled.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Ensure all column names are strings
X.columns = X.columns.astype(str)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Display the scaled training data
X_train_scaled[:5]
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

nn = tf.keras.models.Sequential()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint_callback = ModelCheckpoint(
    filepath='model_weights_epoch_{epoch:02d}.weights.h5',
    save_weights_only=True,
    save_freq='epoch'  # Save after each epoch; you can adjust this based on your needs
)
```

```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam, RMSprop, SGD, Adamax

# Updated Model Architecture
nn = Sequential()
nn.add(Input(shape=(45,)))
nn.add(Dense(units=1512, activation="relu"))
nn.add(Dropout(0.3))  # Adding Dropout to prevent overfitting
nn.add(Dense(units=512, activation="relu"))
nn.add(Dropout(0.3))  # Adding Dropout to prevent overfitting
nn.add(Dense(units=256, activation="relu"))
nn.add(Dense(units=1, activation="sigmoid"))

# Experiment with different optimizers
    #'Adam': Adam(learning_rate=0.001),
    #'RMSprop': RMSprop(learning_rate=0.001),
    #'SGD': SGD(learning_rate=0.01, momentum=0.9),
# Iterate through each optimizer to find the best one: Adamax had best results
optimizers = {
    'Adamax': Adamax(learning_rate=0.002)
}

for name, optimizer in optimizers.items():
    print(f"\nTraining with {name} optimizer...")
    nn.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])

    # Callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)

    # Train the model
    nn.fit(X_train_scaled, y_train, epochs=100, validation_data=(X_test_scaled, y_test),
              callbacks=[early_stopping, reduce_lr], verbose=1)

    # Evaluate the model
    loss, accuracy = nn.evaluate(X_test_scaled, y_test, verbose=0)
    print(f"{name} Optimizer - Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
```

```python
# Export our model to HDF5 file
nn.save('my_model.h5')
```
